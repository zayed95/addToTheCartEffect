### Data Cleaning ###

Standardizing text columns format

```python
df['Major'].unique()
```

```python
# a function to standardize text columns based on specific patterns
def standardize_text(text, mapping):
    text = str(text).strip()
    for pattern, standard in mapping.items():
        if re.search(pattern, text, flags=re.IGNORECASE):
            return standard
    return text
```

```python
mapping = {
    r'languages': 'Arts',
    r'accounting': 'Business',
    r'art': 'Arts',
    r'lit': 'Arts',
    r'ness': 'Business'
}
df['Major'] = df['Major'].apply(lambda x: standardize_text(x, mapping))
standard = ['Computer science', 'Engineering', 'Medical', 'Arts', 'Business']
df['Major'] = df['Major'].apply(lambda x: x if x in standard else 'Other')
df['Major'].unique()
```

```python
df['Platform'].unique()
```

```python
mapping = {
    r'whats': 'WhatsApp',
    r'pin': 'Pinterest',
    r'yotube': 'YouTube'
}
df['Platform'] = df['Platform'].apply(lambda x: standardize_text(x, mapping))
standard = ['YouTube', 'Facebook', 'Twitter/X', 'TikTok', 'Pinterest', 'Instagram']
df['Platform'] = df['Platform'].apply(lambda x: x if str(x).strip() in standard else 'Other')
df['Platform'].unique()
```

```python
df['ImpulsivePurchaseReason'].unique()
```

```python
standard = ['Trendiness', 'Influencer hype', 'Limited-time offers', 'Social proof', 'Ad frequency', 'nan']
df['ImpulsivePurchaseReason'] = df['ImpulsivePurchaseReason'].apply(lambda x: x if str(x).strip() in standard else 'Personal need')
df['ImpulsivePurchaseReason'].unique()
```

Standardizing multiple-answered questions

```python
# Most influncing factor in buying decisions
factors = ['Influencer recommendations', 'Paid ads', 'User reviews', 'Brand posts', 'Viral trends']
influencing_factors = pd.DataFrame(columns=factors)

for factor in factors:
    influencing_factors[factor] = df['ShoppingInfluncedBy'].apply(lambda x: 1 if factor in str(x).strip().split(' ;') else 0)
influencing_factors.head()
```

```python
# Verification Methods before buying a product online
methods = ['Reading reviews', 'Asking friends', 'Checking the offical website', 'Watching YouTube reviews']
verification_methods = pd.DataFrame(columns=methods)

for method in methods:
    verification_methods[method] = df['ProductVerification'].apply(lambda x: 1 if method in str(x).strip().split(';') else 0)
verification_methods.head()
```

```python
# Most influncing factor in gaining customer's trust
trust_factors = ['Real customer reviews', 'Transparency', 'Less aggressive ads', 'Influencer credibility']
trust_factors_df = pd.DataFrame(columns=trust_factors)

for factor in trust_factors:
    trust_factors_df[factor] = df['BrandTrustGainingFactors'].apply(lambda x: 1 if factor in str(x).strip().split(';') else 0)
trust_factors_df.head()
```

```python
# What makes some brands better at marketing than others
quality_factors = ['Quality of ads', 'Trustworthy influencers', 'Strong customer service', 'Consistent social media presence', 'Engaging storytelling']
quality_factors_df = pd.DataFrame(columns=quality_factors)

for factor in quality_factors:
    quality_factors_df[factor] = df['MarketingQualityFactors'].apply(lambda x: 1 if factor in str(x).strip().split(' ;') else 0)
influencing_factors.head()
```

Standardizing the most influential brand on social media

```python
def clean_text(text):
    if pd.isna(text) or str(text).strip().lower() in ['n/a', 'na', 'none']:
        return None

    text = str(text).strip().lower()
    text = re.sub(r'\(.*?\)', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.strip()
    return text if text else None
df['Brand'] = df['Brand'].apply(clean_text)
```

```python
mapping = {
    r'laban': 'B.laban',
    r'لبن': 'B.laban',
    r'labn': 'B.laban',
    r'blbn': 'B.laban',
    r'blban': 'B.laban',
    r'shein': 'Shein',
    r'fatis': "Fati's",
    r'parkville': 'Parkville'
}
df['Brand'] = df['Brand'].apply(lambda x: standardize_text(x, mapping))
standard = ['B.laban', "Fati's", 'Parkville', 'Shein', 'breadfast', 'None']
df['Brand'] = df['Brand'].apply(lambda x: x if str(x).strip() in standard else 'Other')

df['Brand'].value_counts(dropna=True).head(10)
```

### Correlation Analysis ###

```python
# Converting categorical columns to numerical columns
time_mapping = {
    'Less than 1 hour ': 0.5,
    '1-3 hours ': 2,
    '3-5 hours ': 4,
    'More than 5 hours ': 7
}

df['TimeSpentDaily'] = df['TimeSpentDaily'].map(time_mapping)

budget_mapping = {
    'Less than 500 EGP )': 250,
    '500 - 1000 EGP ': 750,
    '1000 - 2000 EGP ': 1500,
    '2000 - 5000 EGP ': 3500,
    'More than 5000 EGP ': 6000
}
df['MonthlyBudget'] = df['MonthlyBudget'].map(budget_mapping)

ordinal_encoder = OrdinalEncoder()
df['ImpulsivePurchase'] = ordinal_encoder.fit_transform(df[['ImpulsivePurchase']])

encoder = OrdinalEncoder(categories=[["rarely", "occasionally", "regularly", "frequently", "very frequently"]])
df['OnlineShoppingFrequency'] = encoder.fit_transform(df[['OnlineShoppingFrequency']])
```
